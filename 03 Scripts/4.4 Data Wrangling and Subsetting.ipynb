# 01 Importing Libraries and Dataframes

```python
#Importing libraries
import pandas as pd
import numpy as np
import os
```

```python
#Importing dataframes
path = r'/Users/Nick/Documents/CareerFoundry/Immersion 4/Instacart Basket Analysis 11.11.21/02 Data/Original Data'
df_ords = pd.read_csv(os.path.join(path, 'orders.csv'), index_col = False)
df_prods = pd.read_csv(os.path.join(path, 'products.csv'), index_col = False)
```

# 02 Data Wrangling

## 02.01 Dropping Columns

```python
#Drop eval_set from Orders dataframe
df_ords = df_ords.drop(columns = ['eval_set'])
```

```python
#Value_count of Orders dataframe
df_ords['days_since_prior_order'].value_counts(dropna = False)
```

### There are 206,209 NaN values.

## 02.02 Renaming Columns

```python
#Change order_dow to orders_day_of_week in Orders dataframe
df_ords.rename(columns = {'order_dow' : 'order_day_of_week'}, inplace = True)
df_ords.head()
```

## 02.03 Changing Variable's Data Types

```python
#Change Orders order_id to a string
df_ords['order_id'] = df_ords['order_id'].astype('str')
df_ords['order_id'].dtype
```

```python
#Change Orders user_id to a string
df_ords['user_id'] = df_ords['user_id'].astype('str')
df_ords['user_id'].dtype
```

## 02.04 Transposing Data

```python
#Import Departments dataframe
df_dep = pd.read_csv(os.path.join(path, 'departments.csv'), index_col = False)
```

```python
#Departments dataframe
df_dep.head()
```

```python
#Trasposing Departments dataframe
df_dep_T = df_dep.T
df_dep_T
```

```python
#Create Index for Departments
df_dep_T.reset_index()
```

```python
#Create header for Departments
new_header = df_dep_T.iloc[0]
#New Departments without header
df_dep_T_new = df_dep_T[1:]
#Set header for Departments
df_dep_T_new.columns = new_header
#Print
df_dep_T_new
```
